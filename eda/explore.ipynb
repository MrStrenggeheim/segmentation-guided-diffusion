```python
from utils.amos import AmosDataset
from torchvision import transforms

dataset_train = AmosDataset(
        "/vol/aimspace/projects/practical_WS2425/diffusion/data/amos_robert_slices/images_all_axis",
        "/vol/aimspace/projects/practical_WS2425/diffusion/data/amos_robert_slices/labels_all_axis",
        "train",
        transform=transforms.Compose(
            [
                transforms.ToTensor(),
                transforms.RandomRotation(10),
                transforms.Resize(256),
                transforms.CenterCrop(256),
                transforms.Normalize(
                    1 * [0.5], 1 * [0.5]
                ),
            ]
        ),
    )
```

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
img_np = np.array(dataset_train[10]['images_target'])[0]
plt.imshow(img_np, cmap='gray')
```

```python
noise_np = np.random.normal(0, .5, img_np.shape)
noised_img_np = img_np + noise_np

plt.imshow(noised_img_np, cmap='gray')
```
